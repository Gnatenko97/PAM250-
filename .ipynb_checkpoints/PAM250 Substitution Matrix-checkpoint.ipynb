```python
import pandas as pd
import numpy as np
```

```python
'''
The Goal of this proggram is to recreate method by which substitution matrix PAM250 was created based on number of amino 
acids substitutions of five closely related bacterial strain domains. 

1. Escherichia coli str. K-12 substr. MG1655
2. Geobacter sulfurreducens PCA
3. Desulforudis audaxviator MP104C
4. Zymomonas mobillis subsp. mpbilis ZM4
5. Bacillus subtilis subsp. subtilis str. 168

  The purpose of this program is to calculate Sij = log2(Mij/Eij), for one 
pair of amino acids, i and j.  Sij is one value in the substitution matrix.
Examile the heat map for Sii and Sjj with PAM250.
'''
#Setp 1 Defining amino acids and opening file 
aa='ARNDCQEGHILKMFPSTWYV'
naa=len(aa)

infile = open("alingnments_2.txt")  #opening allingment file of EMBOSS Needle
countYM = 0
lines = list(infile)  #separate file into lines

m=[[0 for x in range(naa)]for y in range(naa)]
countpairs=0

for i in range(0, len(lines), 4) :  #for all the lines
  a = lines[i]
  b = lines[i+1]
  c = lines[i+2]
  d = lines[i+3]

  if len(d)>5:
    print(d, '  trouble at line ',str(i))
#Go accross line b which tells us where the allingment happens 
  for j in range(len(b)):
        if b[j] in '.:|':
            countpairs +=1
            letterUp,letterDown=a[j],c[j]
            x,y=aa.find(letterUp), aa.find(letterDown) #x and y are position num in aa
            m[x][y]+=1
            m[y][x]+=1
print ('The initial counts of each kind of substitution')
print (aa)
print('-------------------')
for y in range(naa):
    print(m[y]) #number of substitutions say Y for Y would count as 2 (order does matter)
print('----------------')
print('Number of alingment positions:',countpairs)

```

```python
countaa=[0 for i in range(naa)]
for j in range(naa):
    for i in range(naa):
        countaa[i]+=m[i][j] #count how many aa in the seqence alignments not included
        if i==j:                   #idk what this is for
            countaa[i]+=m[i][i]    #idk what this is for 
            
print('The raw count of each amino acid')
print (aa)
print(countaa)
print('------------------------')
```

```python
#Add 1 to the substitution matix 
print(aa) 
print('--------------')

for i in range(naa):  #countaa look above
    countaa[i]=countaa[i]+1;
    for j in range(naa):
        m[i][j]=m[i][j]+1;
print('Adjusted substitution counts')
for i in range(naa):
    print(m[i])

print('-------------------')
print('Number of alingment positions',countpairs)
print('---------------')
print('Adjusted amino acid counts')
print(countaa)
```

```python
countpositions=2*countpairs 
print('Number of amino acid positions',countpositions)
print('--------------------------------------------')
```

```python
freqencyaa=[0 for i in range(naa)]  #calculate freqency of amino acids substitutions 
for i in range(naa):
    freqencyaa[i]=float(countaa[i])/countpositions
    for j in range(naa):
        m[i][j]=float(m[i][j])/countpairs
print('Freqency of substitutions:')

print(aa)
for i in range(naa):
    print('%5.3f'%freqencyaa[i]),
print()
print('-------------')
        
        
```

```python
Ey=[[0 for x in range(naa)]for y in range(naa)] #define E matrix and S matrix 
S=[[0 for x in range(naa)]for y in range(naa)]
```

```python
for i in range(naa):  #fill in freqency matrix 
    for j in range(naa):
        Ey[i][j]=2*freqencyaa[j]*freqencyaa[i]
```

```python
for i in range(naa):   #fill in -odds for actual substitution of j for i, or vice vera
    for j in range(naa):
        S[i][j]=np.log2(m[i][j]/Ey[i][j])
       
```

```python
print("   ",end=" ")  #print matrix
for k in range(naa):
    print(aa[k],end=" ")
    print('   ',end=" ") 
for j in range(naa):
    print()
    print(aa[j],end=" ")
    for i in range(naa):
        print('%5.2f'%S[j][i],end=" ")
```
